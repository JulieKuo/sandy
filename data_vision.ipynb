```python
import matplotlib.pyplot as plt  # Matplotlib 主套件，用來繪製各種圖表
import seaborn as sns  # Seaborn 套件，基於 Matplotlib 提供更多高階統計圖表
import numpy as np  # Numpy 用來產生數據
import pandas as pd  # Pandas 用來處理資料
```

```python
# 設定 Seaborn 的樣式
sns.set(style="whitegrid")
```

```python
# 產生一些隨機數據
np.random.seed(42)  # 固定隨機數據產生器的種子，讓每次運行結果一樣
data = np.random.randn(1000)  # 產生 1000 筆標準常態分布的數據
```

# 畫圖方法1. matplotlib

### 1. 直方圖 (Histogram) - 顯示數據的分布情形

```python
plt.figure(figsize=(8, 6))  # 設定圖表大小
plt.hist(data, bins=30, color='skyblue', edgecolor='black')  # 使用 Matplotlib 繪製直方圖
plt.title('Histogram of Random Data')  # 設定標題
plt.xlabel('Value')  # 設定 X 軸標籤
plt.ylabel('Frequency')  # 設定 Y 軸標籤
plt.show()  # 顯示圖表
```

### 2. 點散圖 (Scatter Plot) - 顯示兩組變數之間的關係

```python
# 產生兩組數據
x = np.linspace(0, 10, 100)  # 產生 0 到 10 之間的 100 個點
y = np.sin(x) + 0.1 * np.random.randn(100)  # 產生有噪音的正弦波

plt.figure(figsize=(8, 6))  # 設定圖表大小
plt.scatter(x, y, color='red')  # 使用 Matplotlib 繪製點散圖
plt.title('Scatter Plot of Sin Function with Noise')  # 設定標題
plt.xlabel('X')  # 設定 X 軸標籤
plt.ylabel('Y')  # 設定 Y 軸標籤
plt.show()  # 顯示圖表
```

### 3. 折線圖 (Line Plot) - 顯示變數隨時間或順序的變化

```python
plt.figure(figsize=(8, 6))  # 設定圖表大小
plt.plot(x, np.sin(x), label='Sine')  # 繪製正弦波折線
plt.plot(x, np.cos(x), label='Cosine', linestyle='--')  # 繪製餘弦波折線，並使用虛線
plt.title('Line Plot of Sine and Cosine Functions')  # 設定標題
plt.xlabel('X')  # 設定 X 軸標籤
plt.ylabel('Y')  # 設定 Y 軸標籤
plt.legend()  # 顯示圖例，區分 Sine 和 Cosine
plt.show()  # 顯示圖表
```

# 畫圖方法2. Seaborn

### 1. 直方圖 和 核密度估計(kde)

```python
plt.figure(figsize=(8, 6))  # 設定圖表大小
sns.histplot(data, bins=30, kde=True, color='blue')  # 使用 Seaborn 繪製直方圖，並顯示核密度估計
plt.title('Seaborn Histogram with KDE')  # 設定標題
plt.xlabel('Value')  # 設定 X 軸標籤
plt.ylabel('Frequency')  # 設定 Y 軸標籤
plt.show()  # 顯示圖表
```

### 2. 點散圖

```python
df = pd.DataFrame({'X': x, 'Y': y})  # 建立 Pandas DataFrame 來存放數據
plt.figure(figsize=(8, 6))  # 設定圖表大小
sns.scatterplot(x='X', y='Y', data=df, color='green')  # 使用 Seaborn 繪製點散圖
plt.title('Seaborn Scatter Plot')  # 設定標題
plt.xlabel('X')  # 設定 X 軸標籤
plt.ylabel('Y')  # 設定 Y 軸標籤
plt.show()  # 顯示圖表
```

### 3. 折線圖

```python
plt.figure(figsize=(8, 6))  # 設定圖表大小
sns.lineplot(x=x, y=np.sin(x), label='Sine')  # 使用 Seaborn 繪製正弦波折線
sns.lineplot(x=x, y=np.cos(x), label='Cosine', linestyle='--')  # 使用 Seaborn 繪製餘弦波折線
plt.title('Seaborn Line Plot of Sine and Cosine')  # 設定標題
plt.xlabel('X')  # 設定 X 軸標籤
plt.ylabel('Y')  # 設定 Y 軸標籤
plt.legend()  # 顯示圖例
plt.show()  # 顯示圖表
```
